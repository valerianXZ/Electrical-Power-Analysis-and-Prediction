```python
def series_to_supervised(data, n_in=7, n_out=1, dropnan=True):
    n_vars = 1 if isinstance(data, list) else data.shape[1]
    df = pd.DataFrame(data)
    cols, names = [], []

    for i in range(n_in, 0, -1):
        cols.append(df.shift(-i))
        names += [f'var{j+1}(t-{i})' for j in range(n_vars)]

    for i in range(n_out):
        cols.append(df.shift(-i))
        if i == 0:
            names += [f'var{j+1}(t)' for j in range(n_vars)]
        else:
            names += [f'var{j+1}(t+{i})' for j in range(n_vars)]
  
    agg = pd.concat(cols, axis=1)
    agg.columns = names

    # Add rolling mean as a new feature
    agg['rolling_mean'] = df.mean(axis=1).shift(-1)
    
    if dropnan:
        agg.dropna(inplace=True)

    return agg
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd

# Load data

df.set_index('Date_time', inplace=True)
df = df.apply(pd.to_numeric, errors='coerce')
df_resample = df.resample('D').mean()

df_resample.shape
values = df_resample.values  
df_resample.to_csv('df_resample.csv')
np.savetxt('values.csv', values, delimiter=',')


scaler = MinMaxScaler(feature_range=(0, 1))
scaled = scaler.fit_transform(values)


reframed = series_to_supervised(scaled, n_in=7, n_out=1)
reframed.to_csv("reframe.csv")
drop_cols = [col for col in reframed.columns if 't+' in col]  
reframed.drop(columns=drop_cols, inplace=True)


values = reframed.values
n_train_time = int(0.8 * len(values))  
train = values[:n_train_time, :]
test = values[n_train_time:, :]

print(f'Train set shape: {train.shape}')
print(f'Test set shape: {test.shape}')
train_x, train_y = train[:, :-1], train[:, -1]
test_x, test_y = test[:, :-1], test[:, -1]


train_x = train_x.reshape((train_x.shape[0], 1, train_x.shape[1]))
test_x = test_x.reshape((test_x.shape[0], 1, test_x.shape[1]))


```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization, Bidirectional
from tensorflow.keras.optimizers import Adam

model = Sequential()
model.add(Bidirectional(LSTM(128, input_shape=(train_x.shape[1], train_x.shape[2]), return_sequences=True)))
model.add(Dropout(0.4))
model.add(BatchNormalization())

model.add(Bidirectional(LSTM(64, return_sequences=True))) #Adding another LSTM layer.
model.add(Dropout(0.3))
model.add(BatchNormalization())

model.add(LSTM(32, return_sequences=False))
model.add(Dropout(0.2))

model.add(Dense(32, activation='relu'))
model.add(Dense(1))

optimizer = Adam(learning_rate=0.001)  
model.compile(loss='mean_squared_error', optimizer=optimizer)


```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = model.fit(train_x, train_y, epochs=100, batch_size=32, validation_data=(test_x, test_y), verbose=2, shuffle=False, callbacks=[early_stopping])
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# Predict
yhat = model.predict(test_x)

# Reshape for inverse scaling
test_x = test_x.reshape((test_x.shape[0], test_x.shape[2]))

# Invert scaling for forecast
inv_yhat = np.concatenate((yhat, test_x[:, -6:]), axis=1)
inv_yhat = scaler.inverse_transform(inv_yhat)
inv_yhat = inv_yhat[:,0]

# Invert scaling for actual values
test_y = test_y.reshape((len(test_y), 1))
inv_y = np.concatenate((test_y, test_x[:, -6:]), axis=1)
inv_y = scaler.inverse_transform(inv_y)
inv_y = inv_y[:,0]

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(inv_y, inv_yhat))
print(f'Test RMSE: {rmse:.3f}')

# Plot loss curves
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
plot_range = 279  # Set the range you want to plot

# Define x-axis values
aa = [x for x in range(plot_range)]

# Ensure we correctly slice both actual and predicted values
plt.figure(figsize=(12, 6))
plt.plot(aa, inv_y[:plot_range], marker='.', label="Actual")  # Plot actual values
plt.plot(aa, inv_yhat[:plot_range], 'r', label="Prediction")  # Plot predicted values
plt.ylabel('Global_active_power', size=15)
plt.xlabel('Time step', size=15)
plt.legend(fontsize=15)
plt.title("Actual vs Predicted Global Active Power", size=16)
plt.show()
```

```python
from sklearn.metrics import r2_score

# Calculate R² score (coefficient of determination)
r2 = r2_score(inv_y, inv_yhat)

# Print R² score
print(f'R² Score: {r2:.4f}')
```

```python
mape = np.mean(np.abs((inv_y - inv_yhat) / inv_y)) * 100

# Print MAPE (mean percentage error)
print(f'MAPE: {mape:.2f}%')
```
